### Funcion que permite leer todos los archivos de una misma extension

```python
def find_files_with_extension(extension):
    if not extension.startswith('.'):
        extension = '.' + extension    
    current_dir = os.getcwd()
    search_pattern = os.path.join(current_dir, f'*{extension}')    
    files = glob.glob(search_pattern)    
    return files
```

### Funcion que permite clasificar las columnas en categoricas, discretas y continuas

```python
def get_variables_scale_type(dataset):  
    columnas = dataset.columns
    categoricas = []
    continuas = []
    discretas = []

    for col in columnas:
        col_type=dataset[col].dtype
        
        if(col_type == 'object' or col_type == 'category'):
            categoricas.append(col)
        elif((col_type =='int64' or col_type =='int32') or (col_type =='float64' or col_type =='float32')):
            n = len(dataset[col].unique())
            if(n > 30):
                continuas.append(col)
            else:
                discretas.append(col)
    
    return continuas, discretas, categoricas       
```

### Función que muestra la información del dataframe

```python
def printDataSetInfo(dataset, excludeColumn, graph = True):
    print(f"{'*'*50} DataSet Information {'*'*50}")
    print(f"Dimensions: {dataset.shape}")
    print(dataset.dtypes)
    print()
    columnas = dataset.columns
    for col in columnas:
        if col in excludeColumn:
            continue
        print(f"{'*'*30} {col} {'*'*30}")        
        print(dataset[col].describe())
        df_describe = dataset[col].describe() 
        unique_values = dataset[col].nunique()
        total_values = dataset[col].count()
        col_type=dataset[col].dtype
        print(f"Column: {col}, Unique Values: {unique_values}, Total Values: {total_values}")      
        if col_type == 'object' and unique_values == total_values:   
            print(f"We are unable to show the distribution graph for {col} due to the amount of values")          
        else:
            if graph == True:
                sns.displot(dataset[col])
                plt.show()
        print(f"{'*'*30} {col} {'*'*30}")          
        print()
```

# Funcion que rellena los valores nulos

```python
def handleNAValues(dataset):    
    columnas = dataset.columns
    for col in columnas:        
        porcentaje = dataset[col].isnull().mean()
        print(f"Percentage of null values for {col}: {porcentaje}%")       
        if porcentaje > 0:   
            dataset[col] = dataset[col].fillna(dataset[col].mean())          
        
    return dataset
```

### Funcion que permite graficar los boxplots

```python
def graphBoxPlot(data, title, graph = True):
    if graph == True:
        fig = plt.figure(figsize =(8, 4))
        plt.title(f"{title}")
        plt.boxplot(data)
        plt.show()
    
    min_val = np.min(data)
    q1 = np.percentile(data, 25)
    median = np.median(data)
    q3 = np.percentile(data, 75)
    max_val = np.max(data)
    print(f'Min: {min_val}')
    print(f'Q1: {q1}')
    print(f'Median: {median}')
    print(f'Q3: {q3}')    
    print(f'Max: {max_val}')
```

### Reemplaza valores invalidos

```python
def replace_invalid_float_values(df, column_name):
    def is_invalid(value):
        try:
            float(value)  
            return False  
        except (ValueError, TypeError):
            return True 
   
    df[column_name] = df[column_name].apply(lambda x: np.nan if is_invalid(x) else float(x))
    mean_value = df[column_name].mean()
    df[column_name].fillna(mean_value, inplace=True)
    
    return df
```

### Obtener metricas de rendimiento de los modelos

```python
def metricsForModels(y_test, y_pred, x_test, model):
    conf_matrix_gnb = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(conf_matrix_gnb)

    precision_gnb = precision_score(y_test, y_pred, average='weighted')
    recall_gnb = recall_score(y_test, y_pred, average='weighted')
    f1_gnb = f1_score(y_test, y_pred, average='weighted')    
    
    y_test_array = np.array(y_test).ravel()
    y_test_proba = model.predict_proba(x_test)

    if y_test_proba.shape[1] > 2:  # Multi-class case
        roc_auc_gnb = roc_auc_score(y_test_array, y_test_proba, multi_class='ovo')
    else:  # Binary case
        roc_auc_gnb = roc_auc_score(y_test_array, y_test_proba[:, 1])
    
    #roc_auc_gnb = roc_auc_score(y_test, y_test_proba, multi_class='ovo')
    print(f"Precision: {precision_gnb}, Recall: {recall_gnb}, F1 Score: {f1_gnb}, AUC-ROC: {roc_auc_gnb}")
```

### Obtener resultados de validacion cruzada

```python
def crossModelValidation(model, x, y, title):
    cv_scores = cross_val_score(model, x, y, cv=CV, scoring='accuracy')
    print(f"Cross-validation scores for {title}:", cv_scores)
    print("Mean cross-validation score:", np.mean(cv_scores))
```

### Trazar curvas ROC para cada modelo

```python
def plot_roc_curve(model, X_test, y_test, model_name):
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)
    elif hasattr(model, "decision_function"):
        y_prob = model.decision_function(X_test)
    else:
        raise AttributeError("Model does not have predict_proba or decision_function method")
    
    fpr = {}
    tpr = {}
    roc_auc = {}
    for i in range(2):
        fpr[i], tpr[i], _ = roc_curve(y_test, y_prob[:, i], pos_label=i)
        roc_auc[i] = auc(fpr[i], tpr[i])

    for i in range(2):
        plt.plot(fpr[i], tpr[i], label=f'ROC curve of class {str(i)} (area = {roc_auc[i]:0.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic for {model_name}')
    plt.legend(loc="lower right")
    plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_complete[continuas])

pca = PCA()
pca_result = pca.fit_transform(scaled_data)

loadings = pca.components_

loading_df = pd.DataFrame(loadings.T, columns=[f'PC{i+1}' for i in range(len(loadings))], index=continuas)

print("Loading matrix (principal component loadings):")
print(loading_df)

most_influential_columns = loading_df['PC1'].abs().sort_values(ascending=False).index.tolist()

print("Most influential columns based on the first principal component:")
print(most_influential_columns)


N = 10 
top_influential_columns = most_influential_columns[:N]
print(f"Top {N} most influential columns:")
print(top_influential_columns)
```

```python
from sklearn.decomposition import PCA

X = df_complete[categoricas].drop(columns=['ID','Customer_ID'])
y = df_complete['Credit_Score']

pca = PCA(n_components=18) 
X_pca = pca.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=RANDOM_STATE)
model.fit(X_train, y_train)
```

```python
importances = model.feature_importances_

pca_importance_df = pd.DataFrame({
    'Principal_Component': [f'PC{i+1}' for i in range(X_pca.shape[1])],
    'Importance': importances
})

pca_importance_df = pca_importance_df.sort_values(by='Importance', ascending=False)
print(pca_importance_df)

pca_loadings = pd.DataFrame(pca.components_.T, index=X.columns, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

top_pcs = pca_importance_df['Principal_Component'].tolist()
top_loadings = pca_loadings[top_pcs]
feature_importance = top_loadings.abs().sum(axis=1)
feature_importance = feature_importance.sort_values(ascending=False)

print(feature_importance)
df2 = feature_importance.reset_index()
feature_importance  = df2['index'].to_list()
print(feature_importance)

```

# 4. Modelo de Deep Learning

```python
X = df_complete[combined_influential_columns]
y = df_complete['Credit_Score']

X_train_val, X_test, y_train_val, y_test = train_test_split(X,  y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=VALIDATE_SIZE, random_state=RANDOM_STATE, stratify=y_train_val)
```

```python
y_train = to_categorical(y_train,num_classes=3)
y_test = to_categorical(y_test,num_classes=3)
y_val =  to_categorical(y_val,num_classes=3)
```

```python
y_train.shape
```

```python
X_train.shape
```

```python
# Configuraicones
configurations = [
    [ Dense(15, input_dim=X_train.shape[1], activation='relu'),
        Dense(15, activation='relu'),
        Dense(3, activation='softmax')],
     [ Dense(15, activation='relu'),
        Dense(15, activation='relu'),
        Dense(3, activation='softmax')],
    [ Dense(18, activation='relu'),
        Dense(18, activation='relu'),
        Dense(3, activation='softmax')]
]
```

```python
def build_model(X_train, config):
    model = Sequential(config)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])    
    return model
```

```python
i = 0
for config in configurations:
    print(f'{"*"*20}{i}{"*"*20}')
    model_1 = build_model(X_train,config)
    history_1 = model_1.fit(X_train, y_train, validation_split=VALIDATE_SIZE, epochs=50, batch_size=32, callbacks=[EarlyStopping(patience=5)])
    loss_1, accuracy_1 = model_1.evaluate(X_val, y_val)
    print(f'Model 1 - Loss: {loss_1}, Accuracy: {accuracy_1}')
    
    plt.plot(history_1.history['accuracy'])
    plt.plot(history_1.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train','Validation'], loc='upper left')
    plt.show()
    i += 1
```

#### Arquitectura Ganador
- Primer Capa: Densa con 18 neuronas que correspondne a los features de mayor importancia.
- Segunda Capa: Densa con 18 neuronas que correspondne a los features de mayor importancia.
- Tercer Capa: Densa con 3 neuronas que corresponde a los valores de clasificacion 0|1|2

```python
model_1 = build_model(X_train, configurations[2])
history_1 = model_1.fit(X_train, y_train, validation_split=0.2, epochs=50, batch_size=32, callbacks=[EarlyStopping(patience=5)])
loss_1, accuracy_1 = model_1.evaluate(X_val, y_val)
print(f'Model 1 - Loss: {loss_1}, Accuracy: {accuracy_1}')
```
