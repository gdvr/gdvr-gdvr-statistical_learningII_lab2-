### Funcion que permite leer todos los archivos de una misma extension

```python
def find_files_with_extension(extension):
    if not extension.startswith('.'):
        extension = '.' + extension    
    current_dir = os.getcwd()
    search_pattern = os.path.join(current_dir, f'*{extension}')    
    files = glob.glob(search_pattern)    
    return files
```

### Funcion que permite clasificar las columnas en categoricas, discretas y continuas

```python
def get_variables_scale_type(dataset):  
    columnas = dataset.columns
    categoricas = []
    continuas = []
    discretas = []

    for col in columnas:
        col_type=dataset[col].dtype
        
        if(col_type == 'object' or col_type == 'category'):
            categoricas.append(col)
        elif((col_type =='int64' or col_type =='int32') or (col_type =='float64' or col_type =='float32')):
            n = len(dataset[col].unique())
            if(n > 30):
                continuas.append(col)
            else:
                discretas.append(col)
    
    return continuas, discretas, categoricas       
```

### Función que muestra la información del dataframe

```python
def printDataSetInfo(dataset, excludeColumn, graph = True):
    print(f"{'*'*50} DataSet Information {'*'*50}")
    print(f"Dimensions: {dataset.shape}")
    print(dataset.dtypes)
    print()
    columnas = dataset.columns
    for col in columnas:
        if col in excludeColumn:
            continue
        print(f"{'*'*30} {col} {'*'*30}")        
        print(dataset[col].describe())
        df_describe = dataset[col].describe() 
        unique_values = dataset[col].nunique()
        total_values = dataset[col].count()
        col_type=dataset[col].dtype
        print(f"Column: {col}, Unique Values: {unique_values}, Total Values: {total_values}")      
        if col_type == 'object' and unique_values == total_values:   
            print(f"We are unable to show the distribution graph for {col} due to the amount of values")          
        else:
            if graph == True:
                sns.displot(dataset[col])
                plt.show()
        print(f"{'*'*30} {col} {'*'*30}")          
        print()
```

# Funcion que rellena los valores nulos

```python
def handleNAValues(dataset):    
    columnas = dataset.columns
    for col in columnas:        
        porcentaje = dataset[col].isnull().mean()
        print(f"Percentage of null values for {col}: {porcentaje}%")       
        if porcentaje > 0:   
            dataset[col] = dataset[col].fillna(dataset[col].mean())          
        
    return dataset
```

### Funcion que permite graficar los boxplots

```python
def graphBoxPlot(data, title, graph = True):
    if graph == True:
        fig = plt.figure(figsize =(8, 4))
        plt.title(f"{title}")
        plt.boxplot(data)
        plt.show()
    
    min_val = np.min(data)
    q1 = np.percentile(data, 25)
    median = np.median(data)
    q3 = np.percentile(data, 75)
    max_val = np.max(data)
    print(f'Min: {min_val}')
    print(f'Q1: {q1}')
    print(f'Median: {median}')
    print(f'Q3: {q3}')    
    print(f'Max: {max_val}')
```

### Reemplaza valores invalidos

```python
def replace_invalid_float_values(df, column_name):
    def is_invalid(value):
        try:
            float(value)  
            return False  
        except (ValueError, TypeError):
            return True 
   
    df[column_name] = df[column_name].apply(lambda x: np.nan if is_invalid(x) else float(x))
    mean_value = df[column_name].mean()
    df[column_name].fillna(mean_value, inplace=True)
    
    return df
```

### Obtener metricas de rendimiento de los modelos

```python
def metricsForModels(y_test, y_pred, x_test, model):
    conf_matrix_gnb = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(conf_matrix_gnb)

    precision_gnb = precision_score(y_test, y_pred, average='weighted')
    recall_gnb = recall_score(y_test, y_pred, average='weighted')
    f1_gnb = f1_score(y_test, y_pred, average='weighted')    
    
    y_test_array = np.array(y_test).ravel()
    y_test_proba = model.predict_proba(x_test)

    if y_test_proba.shape[1] > 2:  # Multi-class case
        roc_auc_gnb = roc_auc_score(y_test_array, y_test_proba, multi_class='ovo')
    else:  # Binary case
        roc_auc_gnb = roc_auc_score(y_test_array, y_test_proba[:, 1])
    
    #roc_auc_gnb = roc_auc_score(y_test, y_test_proba, multi_class='ovo')
    print(f"Precision: {precision_gnb}, Recall: {recall_gnb}, F1 Score: {f1_gnb}, AUC-ROC: {roc_auc_gnb}")
```

### Obtener resultados de validacion cruzada

```python
def crossModelValidation(model, x, y, title):
    cv_scores = cross_val_score(model, x, y, cv=CV, scoring='accuracy')
    print(f"Cross-validation scores for {title}:", cv_scores)
    print("Mean cross-validation score:", np.mean(cv_scores))
```

### Trazar curvas ROC para cada modelo

```python
def plot_roc_curve(model, X_test, y_test, model_name):
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)
    elif hasattr(model, "decision_function"):
        y_prob = model.decision_function(X_test)
    else:
        raise AttributeError("Model does not have predict_proba or decision_function method")
    
    fpr = {}
    tpr = {}
    roc_auc = {}
    for i in range(2):
        fpr[i], tpr[i], _ = roc_curve(y_test, y_prob[:, i], pos_label=i)
        roc_auc[i] = auc(fpr[i], tpr[i])

    for i in range(2):
        plt.plot(fpr[i], tpr[i], label=f'ROC curve of class {str(i)} (area = {roc_auc[i]:0.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic for {model_name}')
    plt.legend(loc="lower right")
    plt.show()
```
